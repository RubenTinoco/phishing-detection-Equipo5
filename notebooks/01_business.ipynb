# 01 - Business Understanding

## Proyecto
Detección de Sitios Web Fraudulentos (Phishing)

## Sector e industria
Este proyecto se desarrolla en el contexto de ciberseguridad y fintech, donde la detección temprana de sitios web fraudulentos es crítica para proteger a los usuarios y reducir riesgos operativos y reputacionales para la organización.

## Contexto del problema
El phishing es una técnica de fraude digital en la que se crean sitios web falsos que imitan páginas legítimas para engañar a los usuarios y obtener credenciales, datos bancarios u otra información sensible.

En entornos financieros y digitales, este tipo de amenaza puede generar pérdidas económicas, robo de identidad, afectación de la confianza del cliente y exposición reputacional para la organización.

## Problema de negocio
La organización necesita identificar de manera automática si un sitio web es legítimo o fraudulento antes de que el usuario interactúe con él.

Actualmente, detectar manualmente sitios de phishing no es escalable ni suficientemente rápido para responder al volumen de amenazas. Por ello, se requiere un sistema de clasificación binaria que permita apoyar decisiones preventivas y reducir la exposición de usuarios y sistemas ante URLs maliciosas.

## Objetivo de negocio
Desarrollar un sistema automático de detección de sitios web de phishing que proteja a los usuarios y a la organización, permitiendo bloquear URLs maliciosas antes de que el usuario interactúe con ellas.

## Preguntas de negocio
Este proyecto busca responder preguntas como:

- ¿Es posible identificar automáticamente si un sitio web presenta características de phishing?
- ¿Qué atributos del sitio web ayudan a diferenciar páginas legítimas de páginas fraudulentas?
- ¿Cómo puede utilizarse esta clasificación para prevenir el acceso a sitios maliciosos antes de que ocurra una interacción riesgosa?m

La variable objetivo del proyecto es `Result`, la cual representa la clase a predecir dentro del problema de clasificación binaria.

## Stakeholders
Los principales stakeholders del proyecto son:

- usuarios finales, que necesitan navegar de forma segura
- organización o empresa, que busca reducir fraude y proteger su reputación
- equipos de ciberseguridad, que requieren mecanismos automáticos de detección
- equipos de riesgo o tecnología, que pueden integrar la solución en procesos preventivos

## Impacto esperado
Un sistema de detección temprana de phishing puede aportar beneficios como:

- reducción del riesgo de fraude digital
- disminución de accesos a sitios maliciosos
- protección de credenciales y datos sensibles
- mejora de la confianza del usuario en los canales digitales
- soporte a decisiones automáticas de bloqueo o alerta

## Criterios de éxito del modelo

Para este proyecto, el éxito del modelo no se evaluará solo por su capacidad de clasificar correctamente, sino por su utilidad para prevenir que los usuarios interactúen con sitios web fraudulentos.

Dado que se trata de un problema de phishing, el criterio de éxito debe priorizar un equilibrio entre detectar la mayor cantidad posible de sitios maliciosos y evitar un exceso de bloqueos incorrectos sobre sitios legítimos.

## Métrica principal

La métrica principal seleccionada para este proyecto es el **F1-score**.

Se elige esta métrica porque combina **precision** y **recall** en una sola medida, lo cual es adecuado para problemas donde tanto los falsos positivos como los falsos negativos tienen impacto.

En este caso:

- un **falso negativo** implica no detectar un sitio de phishing
- un **falso positivo** implica marcar como fraudulento un sitio legítimo

Por ello, el F1-score permite evaluar de forma más equilibrada el desempeño del modelo que una métrica general como accuracy.

## Umbral mínimo aceptable

Como criterio inicial del proyecto, se establece que el modelo debe alcanzar al menos un:

- **F1-score >= 0.85**

Este valor se considera un nivel mínimo aceptable para justificar el uso del modelo como apoyo en la detección preventiva de sitios fraudulentos.

## Restricciones del negocio

El modelo debe cumplir además con las siguientes restricciones de negocio:

- minimizar el riesgo de que sitios de phishing no sean detectados
- mantener un nivel razonable de falsas alertas para no afectar innecesariamente la experiencia del usuario
- ser reproducible dentro del entorno definido por el proyecto
- poder integrarse posteriormente en un flujo automático de alerta o bloqueo preventivo

## Interpretación del éxito del proyecto

El proyecto se considerará exitoso si el modelo logra identificar de manera confiable sitios web de phishing con un desempeño mínimo aceptable y si sus resultados pueden ser utilizados como soporte para decisiones preventivas dentro de un entorno de ciberseguridad.
